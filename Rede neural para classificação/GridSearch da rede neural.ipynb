# Etapa de *GridSearch*

Continuação do projeto em que foram aplicadas redes neurais para classificar se um cliente pode ou não adquirir um empréstimo pessoal.

Importando as bibliotecas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import keras 
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import plot_confusion_matrix,classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ParameterGrid
```

Importando os dados.

```python
bank=pd.read_excel('Bank_modelling_marketing.xlsx')
```

```python
X=bank[['Age', 'Experience', 'Income', 'Family', 'CCAvg','Education', 'Mortgage','Securities Account','CD Account',
        'Online', 'CreditCard']]
y=bank['Personal Loan']
```

```python
lista=['Age','Experience','Income','CCAvg','Mortgage']

for i in lista:
  X[i] = (X[i] - X[i].mean())/(X[i].std())
```

```python
def RedeNeural(optimizer, loss, kernel_initializer,activation, neurons):
  modelo1=Sequential()
  modelo1.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer, input_dim=11))
  modelo1.add(Dense(units=neurons, activation=activation, kernel_initializer=kernel_initializer))
  modelo1.add(Dense(units=1,activation='sigmoid'))
  modelo1.compile(optimizer=optimizer,loss=loss, metrics=['binary_accuracy'])

  return modelo1
```

```python
modelo_gs = KerasClassifier(build_fn=RedeNeural)
```

Como o procedimento de *GridSearchCV* é bastante demorado e pode levar horas e em muitas situações pode até travar. Seu eu usar todos os parâmetros abaixo terei que treinar 64 modelos diferentes.

```python
parametros = {'batch_size':[10],
              'epochs':[50,100], 
              'optimizer':['adam','sgd'], 
              'loss':['binary_crossentropy','hinge'],
              'kernel_initializer':['random_uniform','normal'], 'activation':['relu','tanh'], 'neurons':[6,12]}

pg = ParameterGrid(parametros)
len(pg)
```

Por esse motivo irei aplicar o *GridSearchCV* de 2 em dois parâmetros diferentes, e nesse caso serão 4 modelos diferentes por procedimento.


## Variando o número de neurônios (*neurons*)


```python
parametros2 = {'batch_size':[10],
              'epochs':[50], 
              'optimizer':['adam'], 
              'loss':['binary_crossentropy'],
              'kernel_initializer':['random_uniform'], 'activation':['relu'], 'neurons':[6,12]}
```

```python
grid_search = GridSearchCV(estimator=modelo_gs, param_grid=parametros2, scoring='accuracy',cv=5)
grid_search = grid_search.fit(X,y)
```

Melhores parâmetros.
